```python
import matplotlib.pyplot as plt
import numpy as np 

a = np.linspace(0,10,11)
b = a**4

x = np.arange(0,10)
y = x * 2

fig,axes = plt.subplots(nrows=2, ncols=2)

axes[0,1].plot(x,y)
axes[0,0].plot(x,y)
axes[1,1].plot(x,y)
axes[1,0].plot(x,y)

fig.subplots_adjust(wspace=0.9,hspace=1)
fig.suptitle("figure level title")
plt.show()
```
