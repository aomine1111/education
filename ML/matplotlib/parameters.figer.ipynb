```python
import matplotlib.pyplot as plt
import numpy as np 

a = np.linspace(0,10,11)
b = a**4

fig = plt.figure(figsize=(10,3),dpi=100)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Воведение в степень 4")

fig.savefig("1231",bbox_inches="tight")
```
